```python
#chrome driver 설치 코드
def get_chrome_driver():
    #1. 크롬 옵션 세팅
    chrome_options = webdriver.ChromeOptions()
    #2.driver 생성
    driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
)

    return driver
```

## BeautifulSoup를 이용한 아이콘 추출하기

```python
driver=get_chrome_driver()
driver.set_window_size(2000, 1000)
driver.set_window_position(1,0) 
driver.get("https://www.emart24.co.kr/store")
driver.implicitly_wait(1)
store_icon_list=[]
#지역선택
store_dropdown="/html/body/div[2]/div/div/div[3]/div[2]/div[1]/div[1]/div[2]/div[1]/div"
store_select = driver.find_element(By.XPATH, store_dropdown)
time.sleep(1)
#서울특별시 선택
seoul_select="/html/body/div[2]/div/div/div[3]/div[2]/div[1]/div[1]/div[2]/div[1]/ul/li[10]"
seoul_select=driver.find_element(By.XPATH, seoul_select)
store_select.click()
seoul_select.click()

#돋보기 버튼 클릭하여 정렬시킴
button="/html/body/div[2]/div/div/div[3]/div[2]/div[1]/div[1]/div[3]/div/div/div"
button_click=driver.find_element(By.XPATH, button)
button_click.click()
result_html=driver.page_source
action=ActionChains(driver)
time.sleep(1)

for i in range(25):
    next_button='/html/body/div[2]/div/div/div[3]/div[3]/div[2]/div[1]'
    next_button_click=driver.find_element(By.XPATH,next_button)
    next_button_click.click()
    time.sleep(1)

    soup=BeautifulSoup(result_html, 'html.parser')
for data_icon in soup.select('ul.choiceServiceTag > li.tag > div.icon >img'):
    if 'data-icon' in data_icon.attrs:  # alt 속성이 있는지 확인
        store_icon_list.append(data_icon['data-icon'])
    
driver.quit()
```
